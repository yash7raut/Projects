View to Like Ratio

```python
view_to_like = []

data = df.values
for i in range(len(data)): 
    view_to_like.append(round(data[i][3]/data[i][4],2))
```

```python
df['VTL_Ratio'] = view_to_like
```

```python
df.sort_values(by='VTL_Ratio',ascending=False).head(10)
```

Hpw many TED talks are given in 2020

```python
data = df.values

month = []
year = []

for i in range(len(data)):
    month.append(data[i][2].split(' ')[0])
    year.append(data[i][2].split(' ')[1])
```

```python
df['year'] = year
df['month'] = month
```

```python
df.head()
```

```python
df[df['year'] == '2020']
```

Total No. of TED Talks given in 2020

```python
len(df[df['year'] == '2020'])
```

Trend of TED Talks frequency by Month in 2020

```python
df_pr
```

```python
df_pr=df[df['year'] == '2021']

df_pr.groupby('month').size().sort_values(ascending = False ).plot(kind='bar')
```

```python
df_pr=df[df['year'] == '2019']

df_pr.groupby('month').size().sort_values(ascending = False ).plot(kind='bar')
```

After seeing the above three plots we can infer that in a particular year, their is a month which have a large amout of TED talks. This may be beacause some global event might be hosted by the TED Community in that particular month.
