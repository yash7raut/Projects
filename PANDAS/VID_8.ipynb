How do I filter rows of a pandas DataFrame by column value?

```python
import pandas as pd
movies = pd.read_csv('https://bit.ly/imdbratings')
movies.head()
```

```python
movies.shape
```

```python
boolean = []
for length in movies.duration:
    if length >= 200:
        boolean.append(True)
    else:
        boolean.append(False)
```

```python
boolean[0:5]
```

```python
boolean[0:15]
```

```python
len(boolean)
```

```python
is_long = pd.Series(boolean)
is_long.head()
```

```python
is_long = movies.duration >= 200
is_long.head()
```

```python
movies[is_long]
```

```python
movies['genre']
```
