How do I apply multiple filter criteria to a pandas DataFrame?

```python
import pandas as pd
movies = pd.read_csv('https://bit.ly/imdbratings')
movies.head()
```

```python
movies[movies.duration >= 200]
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
movies[(movies.duration <= 90) & (movies.star_rating >= 8.0)]
```

```python
movies[(movies.genre == 'Drama') | (movies.genre == 'Crime') | (movies.genre == 'Action')]
```

this above thing can be done with the isin attribute


```python
movies[movies.genre.isin(['crime', 'Drama', 'Action'])]
```
