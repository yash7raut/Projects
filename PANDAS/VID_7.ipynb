SORTING A PANDAS DATAFRAME OR SERIES - look at the codes carefully - these are interesting ones 


```python
import pandas as pd
movies = pd.read_csv('https://bit.ly/imdbratings')
movies.head()
```

```python
movies.title.sort_values()
```

```python
movies.title.sort_values(ascending=False)
```

```python
movies['title'].sort_values()
```

```python
movies['title']
```

```python
movies.sort_values('title')
```

```python
movies.sort_values('duration')
```

```python
movies.sort_values('duration',ascending=False)
```

```python
movies.head()
```

```python
movies.sort_values(['content_rating','duration'],ascending=False)
```

```python
movies.sort_values(['star_rating','duration'],ascending=False)
```
