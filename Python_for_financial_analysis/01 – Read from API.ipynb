Project – Normalize and Calculate Return of Portfolio

normalization of ratings means adjusting values measured on different scales to a notionally common scale, often prior to averaging.”

```python
data = data['Adj Close']
```

```python
data.head()
```

Adj Close = "The adjusted closing price amends a stock's closing price to reflect that stock's value after accounting for any corporate actions. It is often used when examining historical returns or doing a detailed analysis of past performance."

```python
norm = data/data.iloc[0]
```

```python
norm.head(10)
```

As you see, all the start prices are adjusted to 1. This is the normalization

Then the relative change can be seen from day 2010-01-05 and forward

```python
portfolio = [.25,.25,.25,.25]
```

```python
weights = (norm*portfolio)
```

```python
weights.head(10)
```

we get a weighted distribution of our portfolio. Starting with 25% for each ticker

```python
weights['Total'] = (norm*portfolio).sum(axis=1)
```

```python
weights.head(10)
```

we have the sum of our weighted sum of how the portfolio evolves in the Total column

Let’s assume we invested 100,000 Rupees. This can be modeled quite easy as you see here.

```python
(weights*100000).head(10)
```

Our portfolio looks like this in the end

```python
(weights*100000).tail(10)
```

We have that 2022-03-11 we will have 648956.53 in total.

```python
(weights['Total']*100000).iloc[-1]
```

Hence, this shows that if we invested 100,000 Rupees the opening day of 2010 in a portfolio consisting of 25% of each ticker, then we would have 648956.5305837922 Rupess on March 11th, 2022.
